```python
#In this block of code I import all the libraries I will use below.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

##**Demand**

*Demand* refers to the quantity of a good or service that consumers are willing and able to purchase at different prices during a given period. The *Law of Demand* states that, all else being equal, there is an inverse relationship between price and quantity demanded.

A *demand curve* is a graphical representation of the relationship between the price of a good and the quantity demanded. It typically slopes downward to reflect the Law of Demand. The curve shows the different quantities of the good that consumers would purchase at various price levels.

An example of a linear demand curve for the ice cream market is shown below, illustrating how the  quantity (number of gallons) of ice cream consumers are willing to buy changes as the price per gallon varies.








```python
# Create quantity values (e.g., 0 to 20 gallons)
quantity = np.arange(0, 21, 1)

# Create the demand curve: P = 15 - 0.5Q
demand_price = 15 - 0.5 * quantity

# Plot the demand curve
plt.figure(figsize=(8, 6))
plt.plot(quantity, demand_price, color='blue', label='Demand Curve')
plt.title('Demand Curve for Ice Cream')
plt.xlabel('Quantity Demanded (gallons)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

##**Supply**

*Supply* maps the quantity of a good or service that producers are willing and able to sell at different prices during a given period. The *Law of Supply* states that, all else being equal, there is a positive relationship between price and quantity supplied.

A supply curve is a graphical representation of the relationship between the price of a good and the quantity supplied. It typically slopes upward, reflecting the Law of Supply.

An example of a linear supply curve for the ice cream market is shown below, illustrating how the quantity (number of gallons) of ice cream producers are willing to supply changes as the price per gallon varies.

```python
# Create the supply curve: P = 5 + 0.5Q
supply_price = 5 + 0.5 * quantity

# Plot the supply curve
plt.figure(figsize=(8, 6))
plt.plot(quantity, supply_price, color='red', label='Supply Curve')
plt.title('Supply Curve for Ice Cream')
plt.xlabel('Quantity Supplied (gallons)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

##**Market Equilibrium**

*Market equilibrium* occurs when the quantity of a good or service demanded by consumers equals the quantity supplied by producers at a specific price. At this point, the market is balanced, with no excess supply or demand, and the price tends to remain stable.

Incentives play a crucial role in driving the market toward equilibrium by responding to *surpluses* and *shortages*:

Surplus: When the price is above the equilibrium level, the quantity supplied exceeds the quantity demanded, creating a surplus. Producers are incentivized to lower their prices to sell off the excess inventory. As prices fall, the quantity demanded increases and the quantity supplied decreases, moving the market back toward equilibrium.

Shortage: When the price is below the equilibrium level, the quantity demanded exceeds the quantity supplied, resulting in a shortage. Consumers, facing scarcity, are willing to pay higher prices to obtain the product. Producers, seeing an opportunity for higher profits, are incentivized to increase their prices and supply more. This rise in prices reduces the quantity demanded and increases the quantity supplied, again guiding the market toward equilibrium.

These incentives ensure that any imbalances in the marketwhether a surplus or shortagecreate pressure to adjust prices and quantities, leading the market back towards equilibrium where supply matches demand.

Below, we plot both supply and demand in order to observe the ice cream market equilibrium.

```python
# Plot both the demand and supply curves together
plt.figure(figsize=(8, 6))
plt.plot(quantity, demand_price, color='blue', label='Demand Curve')
plt.plot(quantity, supply_price, color='red', label='Supply Curve')
plt.title('Ice Cream Market: Demand and Supply Curves')
plt.xlabel('Quantity (gallons)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

##**Demand Shifts**

A *demand shift* occurs when the quantity demanded of a good or service changes at every price level, leading to a new demand curve. This shift can be either to the right (an increase in demand) or to the left (a decrease in demand). Unlike movements along the demand curve, which are caused by changes in the price of the good itself, demand shifts are caused by changes in other factors, often referred to as *determinants of demand*.

Some general determinants of demand are:

1. Consumer income
2. Consumer preferences
3. Price of related goods: complements and substitutes
4. Expectations of future prices
5. Government policies
6. Natural events

In our ice cream market example, suppose there is a heat wave that increases the price consumers are willing to pay for a fixed quantity of ice cream, or increases the their quantity demanded for a fixed price. Such an event might be shown as an outward shift in demand as shown below.

```python
# New demand curve after outward shift: P = 18 - 0.5Q
demand_price_new = 18 - 0.5 * quantity

# Plot the original and new demand curves with the original supply curve
plt.figure(figsize=(8, 6))
plt.plot(quantity, demand_price, color='blue', label='Original Demand Curve')
plt.plot(quantity, demand_price_new, color='blue', linestyle='--', label='New Demand Curve (Outward Shift)')
plt.plot(quantity, supply_price, color='red', label='Supply Curve')
plt.title('Ice Cream Market: Demand Shift')
plt.xlabel('Quantity (gallons)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

##**Supply Shifts**

A *supply shift* occurs when the quantity supplied of a good or service changes at every price level, resulting in a new supply curve. This shift can either be to the right (an increase in supply) or to the left (a decrease in supply). Unlike movements along the supply curve, which are caused by changes in the price of the good itself, supply shifts are caused by changes in other factors, often referred to as *determinants of supply*.

Some general determinants of supply are:

1. Input prices (costs)
2. Technology
3. Number of producers
4. Expectations of future prices
5. Government policies
6. Natural events

In our ice cream market example, suppose a new type of efficient freezer is invented that leads production costs to decline. This market impact is illustrated below.

```python
# New supply curve after outward shift: P = 2 + 0.5Q
supply_price_new = 8 + 0.5 * quantity

# Plot the original and new supply curves with the original demand curve
plt.figure(figsize=(8, 6))
plt.plot(quantity, demand_price, color='blue', label='Demand Curve')
plt.plot(quantity, supply_price, color='red', label='Original Supply Curve')
plt.plot(quantity, supply_price_new, color='red', linestyle='--', label='New Supply Curve (Outward Shift)')
plt.title('Ice Cream Market: Supply Shift')
plt.xlabel('Quantity (gallons)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

Supply and demand are shifting all of the time for the reasons descrobed above. Let's add a lot of shifts and show the curves on a plot.

```python
# Create quantity values (e.g., 0 to 20 gallons)
quantity = np.arange(0, 21, 1)

# Base slope parameters
demand_slope = -0.5
supply_slope = 0.5

# Plot the demand and supply curves
plt.figure(figsize=(8, 6))

# Equilibrium points storage
equilibrium_points = []

# Loop to create and plot multiple demand and supply curves with increasing and alternating shifts
for i in range(5):
    # Increasing and alternating shifts: magnitude grows with i
    demand_shift = 3 * i * (-1)**i  # Alternates and increases: +0, -3, +6, -9, ...
    supply_shift = 2 * i * (-1)**i  # Alternates and increases: +0, -2, +4, -6, ...

    # Define demand and supply intercepts with shifts
    demand_intercept = 15 + demand_shift
    supply_intercept = 5 + supply_shift

    # Demand curve
    demand_price = demand_intercept + demand_slope * quantity

    # Supply curve
    supply_price = supply_intercept + supply_slope * quantity

    # Plot demand curve
    plt.plot(quantity, demand_price, color='blue', label=f'Demand Curve {i+1}' if i == 0 else "_nolegend_")

    # Plot supply curve
    plt.plot(quantity, supply_price, color='red', label=f'Supply Curve {i+1}' if i == 0 else "_nolegend_")

    # Calculate equilibrium (where demand = supply)
    equilibrium_quantity = (demand_intercept - supply_intercept) / (supply_slope - demand_slope)
    equilibrium_price = demand_intercept + demand_slope * equilibrium_quantity

    # Store equilibrium point
    equilibrium_points.append((equilibrium_quantity, equilibrium_price))

# Add titles and labels
plt.title('Ice Cream Market: Multiple Demand and Supply Curves with Increasing and Alternating Shifts')
plt.xlabel('Quantity (gallons)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()

```

##**Going to the Data**##

Note that when we observe markets, we observe equiibrium outcomes but we don't directly observe the underlying supply and demand curves. For example, if you recorded prices and quantities at a few different ice cream shops in Ogden, you would have a data set of equilibroium prices and equilibrium quantities.

Let's plot some of the equilibrium points from our example above, then draw a line through them to see what happens.

```python
# Extract equilibrium quantities and prices
equilibrium_quantities = np.array([point[0] for point in equilibrium_points]).reshape(-1, 1)
equilibrium_prices = np.array([point[1] for point in equilibrium_points])

# Fit a linear regression model to the equilibrium points
model = LinearRegression().fit(equilibrium_quantities, equilibrium_prices)

# Generate predicted prices along the line of best fit
predicted_prices = model.predict(equilibrium_quantities)

# Get the slope of the regression line
slope = model.coef_[0]

# Plot the equilibrium points
plt.figure(figsize=(8, 6))
plt.scatter(equilibrium_quantities, equilibrium_prices, color='purple', label='Equilibrium Points')

# Plot the line of best fit through the equilibrium points
plt.plot(equilibrium_quantities, predicted_prices, color='green', label='Line of Best Fit')

# Print the slope on the plot, adjusting position to avoid the legend
plt.text(equilibrium_quantities.min() + 0.5, equilibrium_prices.max() - 1, f'Slope: {slope:.2f}', fontsize=12, color='green')

# Add labels and titles
plt.title('Equilibrium Points with Line of Best Fit')
plt.xlabel('Equilibrium Quantity (gallons)')
plt.ylabel('Equilibrium Price ($)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

But slope = 2.5 is not meaningful in our example! That is neither the slope of the demand or supply curve. If we would've used a different subset of equilibria, we would get different slopes that are also unlikely to be the slopes we are interested in. Unless, we critically think about the sample!


# The Value of Econometrics in Data Science

Econometrics is a powerful tool in Data Science because it allows us to uncover the true relationships between variables in complex, real-world situations where traditional data analysis methods might fall short. In the example of supply and demand, econometrics helps us understand that the observed relationship between price and quantity is influenced by simultaneous, interdependent factors. This is not unique to economics; its a fundamental challenge across various fields. By applying econometric techniques, we can disentangle these intertwined relationships, making it possible to derive meaningful insights that drive informed decisions. Whether you're predicting market behavior, evaluating policy impacts, or optimizing business strategies, econometrics equips you with the ability to reason through and solve problems that involve endogeneity and simultaneity, which are common and critical issues in any data-driven field.

# Simultaneity Simulation

```python
# Set seed for reproducibility
np.random.seed(42)

# Number of observations
n = 1000

# True parameters for inverse demand and supply curves
alpha_0 = 50  # Demand intercept
alpha_1 = -2  # Demand slope (negative)
beta_0 = 20  # Supply intercept
beta_1 = 1.5  # Supply slope (positive)

# Generate random shocks (simultaneous shocks affecting both demand and supply)
epsilon_d = np.random.normal(0, 1, n)  # Demand shocks
epsilon_s = np.random.normal(0, 1, n)  # Supply shocks

# Simulate equilibrium quantities and prices
quantities = (beta_0 - alpha_0 + epsilon_d - epsilon_s) / (alpha_1 - beta_1)  # Equilibrium quantity
prices = alpha_0 + alpha_1 * quantities + epsilon_d  # Equilibrium price from inverse demand curve

# OLS Estimation of the Demand Curve (biased because of simultaneity)
X = sm.add_constant(quantities)  # Quantities as the independent variable
y = prices  # Prices as the dependent variable (price = f(quantity))

# Fit OLS
model_biased = sm.OLS(y, X)
results_biased = model_biased.fit()

# Print the OLS results for the full dataset (biased)
print("OLS Results with Simultaneity (Biased):")
print(results_biased.summary())

# Calculate covariance between price (P) and demand shock (_d)
cov_P_epsilon_d = np.cov(prices, epsilon_d)[0, 1]
print(f"\nCovariance between Price (P) and Demand Shock (_d) in Full Dataset: {cov_P_epsilon_d:.4f}\n")

```

```python
# Filter dataset where only supply shifts (we assume epsilon_d is near zero)
supply_only_condition = np.abs(epsilon_d) < 0.1  # Example: small demand shocks
quantities_supply_only = quantities[supply_only_condition]
prices_supply_only = prices[supply_only_condition]

# OLS Estimation of the Demand Curve (unbiased because only supply is shifting)
X_supply_only = sm.add_constant(quantities_supply_only)  # Quantities as independent variable
y_supply_only = prices_supply_only  # Prices as dependent variable

# Fit OLS
model_unbiased = sm.OLS(y_supply_only, X_supply_only)
results_unbiased = model_unbiased.fit()

# Print the OLS results for supply shifts only (unbiased)
print("\nOLS Results with Supply Shifts Only (Unbiased):")
print(results_unbiased.summary())

# Calculate covariance between price (P) and demand shock (_d) for supply shifts only
cov_P_epsilon_d_supply_only = np.cov(prices_supply_only, epsilon_d[supply_only_condition])[0, 1]
print(f"\nCovariance between Price (P) and Demand Shock (_d) in Supply Shifts Only Data: {cov_P_epsilon_d_supply_only:.4f}\n")
```

```python
# Plot the full dataset and highlight the "supply only" subset
plt.figure(figsize=(10, 6))

# Full dataset: Both supply and demand are shifting
plt.scatter(quantities, prices, alpha=0.5, color='blue', label='Full Dataset (Supply & Demand Shifting)')

# "Supply only" subset: Only supply is shifting, demand shocks near zero
plt.scatter(quantities_supply_only, prices_supply_only, alpha=0.8, color='orange', label='Supply Shifts Only', edgecolor='black')

# Add labels, title, and legend
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Simulated Data: Full Dataset vs. Supply Shifts Only')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
